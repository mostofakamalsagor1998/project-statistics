<h1> <font color="red"> Period of a TPM </font> </h1>

```python
import math
import numpy as np

a=np.array([
    [0.2,0.5,0.3,0],#1st row...
    [0.1,0.3,0.6,0],#2nd row...
    [0,0.4,0.3,0.3],#3rd row...
    [0,0,0.5,0.5]#4th row
])
print("\nTransition Probability Matrix(TPM): \n",a)


an=np.dot(a,a)
print(f"\nP^2 : ")
print(an.round(decimals=4))
print()

states=[]
m=1#d(s)=Tss
n=1#d(s)=Tss


# Next occurances
for i in range(3,33):
    an=np.dot(an,a)
    print(f"P^{i} : ")
    print(an.round(decimals=4))
    print()
    if an[m][n]!=0:
        states.append(i)
print(f"\nStates: {states}")

#Gcd 
m=math.gcd(states[0],states[1])
for s in range(2,len(states)):
    m=math.gcd(m,states[s])
print(f"\nGCD: {m}")
```

```python
import numpy as np
import math
import pandas
# import fractions
# np.set_printoptions(formatter={'all':lambda x: str(fractions.Fraction(x))})

P=np.array([
        [0.7,0,0.3,0],#1st row...
        [0.5,0,0.5,0],#2nd row...
        [0,0.4,0,0.6],#3rd row...
        [0,0.2,0,0.8]#4th row
])#row-wise...!!!


an=np.dot(P,P)
t2=an.round(decimals=4)
tn=an.round(decimals=4)


for n in range(3,1025):
    an=np.dot(an,P)
    if n in [4,8,16,32,64,128,512,1024]:
        if an.round(decimals=4).tolist()==tn.tolist():
            print("n = ",n-n//2)
            break
        else:
            tn=an.round(decimals=4)
```
